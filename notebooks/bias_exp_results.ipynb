```python

plt.rcParams.update({'font.size': 16, 'font.family':'serif'})
plt.rcParams["figure.figsize"] = [6.00, 5.0]
plt.rcParams["figure.autolayout"] = True

metrics = ['JAC','DSC','F2','PPV','Rec', 'Acc']
metric_labels = ['JAC','DICE','F2','PPV','Rec', 'Acc']

x_labels = [0, 15, 29, 59, 116]
x_values = [0, 1, 2, 3, 4]


def save_plot(df, title, figname):
    fig, ax = plt.subplots()
    plt.xticks(x_values, x_labels)
    plt.xlim(-0.5, 4.5)
    plt.ylim(0.5, 1)
    plt.grid()

    plt.title(title)
    legends = []
    colors = ['b', 'r', 'g', 'brown', 'purple', 'orange']
    for i, m in enumerate(metrics):
        line, = ax.plot(x_values, df[m], marker="o", markersize=5, color=colors[i])

        legends.append(line)

    leg1 = ax.legend(legends[:3], metric_labels[:3], loc='lower left')
    leg2 = ax.legend(legends[3:], metric_labels[3:], loc='lower right')    

    ax.add_artist(leg1)
    ax.add_artist(leg2)

#     plt.show()
    plt.savefig(figname + ".pdf", format="pdf", bbox_inches="tight") 
```
